### 1. Budget program
Ask user to enter:
- their monthly budget overall, and the monthly budget for each category (i.e., rent, groceries, transportation, etc. Need at least 5 categories). 
- their expenses in the past month for each category (user can choose how many expenses they want to enter for each category)
- monthly income
- which city they are living in
- other things you might want to add (e.g., user's goal for savings)


Then display the following:
1. highest and lowest values for expenses, in general
2. highest and lowest categories of expenses (category names and amounts)
3. the % of total budget each expense category contributed to
4. whether the person exceeded their total monthly budget, by how much, in amount & %
5. whether the person exceeded each budgeted expense category, by how much, in amount & %
6. amount and % of total budget, and of each cateogory's budget left, if any
7. whether the person's rent is above/below the median rent in Manhattan/Brooklyn/Queens/etc.
8. whether the person is rent burdened (more than 30% of income would be considered rent burdened)
9. other features you might want to add (e.g., user's progress in achieving their savings goal, or standard recommendations for savings expressed in % of their income, and amounts).

Make sure all amounts displayed have appropriate format. And that you have validation for all inputs.

```python
import operator
## to prevent a "float division by zero" error later on, I set this error validation to go if the user enters 0 in any budget.
## I will use it for budgets and income
def valid_budget(x):
    while x <= 0:       
        x = float(input("Please enter a positive number."))

## this input validation will be for the expenses        
def valid_expense(x):
    while x < 0:      
        x = float(input("Please enter a positive number."))
        
#________________BUDGETS_______________________________________________________

home_budgets = ['rent', 'electricity', 'gas', 'water', 'phone', 'cable', 'internet', 'appliances', 'other']
home_budgets_cost = []
budget = float(input("Please Enter your monthly budget: "))
valid_budget(budget)

budgets = {}
total_budgets = 0
for i in home_budgets:
    budget_per_cat = float(input("Please enter your budget for " + i + ": "))
    valid_budget(budget_per_cat)
    home_budgets_cost.append(budget_per_cat)
    total_budgets += budget_per_cat
    if i not in budgets:
        budgets[i] = budget_per_cat
    
#____________________EXPENSES__________________________________________________
total_expenses = 0
expenses = {}
for i in home_budgets:
    expense = float(input("\nWhat was your expenses on " + i + " last month? "))
    valid_expense(expense)
    total_expenses += expense
    if i not in expenses:
        expenses[i] = expense
    another_expense = input("Do you want to add another exense for " + i + " ? (enter 'N' if you don't. or enter anything for a 'yes')")
    while another_expense != "N":
        extra_expense = float(input("What was your other expenses on " + i + " last month? "))
        valid_expense(extra_expense)
        another_expense = input("Do you want to add another exense for " + i + " ? (enter 'N' if you don't. or enter anything for a 'yes')")
        expenses[i] += extra_expense
        total_expenses += extra_expense
        
        
#____________________MONTHLY INCOME______________________________________________
income = float(input("\nWhat is your monthly income? "))
valid_budget(income)
        
#____________________CITY________________________________________________________
cities = ['Manhattan', 'Queens', 'Brooklyn', 'Bronx', 'Staten Island']
print("\nHere is a list of the cities in New York:")
for i in cities:
    print(i)
city = input("what city do you live in? ")
while city not in cities:
    print("please type a correct city.")
    city = input("\nWhat city do you live in? ")


#____________________OUTPUT______________________________________________________
sorted_budgets = sorted(budgets.items(), key=operator.itemgetter(1))   ### converted the dictionary into a list of tuples sorted by key value using the operator module (from small to big)
sorted_expenses = sorted(expenses.items(), key=operator.itemgetter(1))  ### converted the dictionary into a list of tuples sorted by key value using the operator module (from small to big)

## 1 & 2
max_expenses = []
M = max(sorted_expenses, key = operator.itemgetter(1))[1]
for i in range(len(sorted_expenses)):
    if sorted_expenses[i][1] == M:
        max_expenses.append(sorted_expenses[i])
        
        
min_expenses = []
N = min(sorted_expenses, key = operator.itemgetter(1))[1]
for i in range(len(sorted_expenses)):
    if sorted_expenses[i][1] == N:
        min_expenses.append(sorted_expenses[i])
        
print("\nThe lowest expenses was in the following categories:")
print("Category\t\tAmount")
for i, k in min_expenses:
    print(i, '\t\t\t $' + str(k))
    
print("\nThe highest expenses was in the following categories:")
print("Category\t\tAmount")
for i, k in max_expenses:
    print(i, '\t\t\t $' + str(k))
    


## 3
perc_list = []
for i in sorted_expenses:
    perc = format((i[1]/total_expenses)*100, '0.2f')
    perc_list.append(i[0] + ": %" + perc)      ### "days" is a list that i created with the weekdays in the beginning
print("\nThe percentage contribution for each category is: ")
for i in perc_list:
    print(i)


## 4
print("\n")
dificit = float(total_expenses - total_budgets)
excess = float(total_budgets - total_expenses)
if total_expenses > total_budgets:
    print("You have exceeded your budget by $" + str(dificit) + 
          " (%" + format(dificit/total_budgets * 100, '0.2f') + ")")
else:
    print("You have a remaining $" + str(excess) + 
          " of your budget (%" + format(excess/total_budgets * 100, '0.2f') + ")\n")
    
    
## 5 & 6
sorted_budgets_key = sorted(budgets.items(), key=operator.itemgetter(0))    ### converted the dictionary again into a list of tuples but sorted by key using the operator module (alphabetically)
sorted_expenses_key = sorted(expenses.items(), key=operator.itemgetter(0))   ### converted the dictionary again into a list of tuples but sorted by key using the operator module (alphabetically)
#print(sorted_expenses_key)      Remove the '#' to double check the math
#print(sorted_budgets_key)
for i in range(9):
    if sorted_expenses_key[i][1] > sorted_budgets_key[i][1]:
        print("You have exceeded your " + sorted_expenses_key[i][0] + " budget by $" 
              + str(sorted_expenses_key[i][1] - sorted_budgets_key[i][1])  + " (%" + 
              str(format((sorted_expenses_key[i][1] - sorted_budgets_key[i][1]) / sorted_budgets_key[i][1] * 100, '0.2f'))
              + ")")
    else:
        print("You have $" + str(sorted_budgets_key[i][1] - sorted_expenses_key[i][1]) + " (%" 
              + str(format((sorted_budgets_key[i][1] - sorted_expenses_key[i][1]) / sorted_budgets_key[i][1] * 100, '0.2f')
              + ") left from your " + sorted_expenses_key[i][0]))
        
## 7
#Manhattan: $4,044 Brooklyn: $3,045 Queens: $2,992 Bronx: $1,700 Staten Island: $1,500 (for reference)        
median_rent = [('Manhattan', 4044), ('Queens', 2992), ('Brooklyn', 3045), ('Bronx', 1700), ('Staten Island', 1500)]
for i in median_rent:
    if i[0] == city:
        if sorted_budgets_key[7][1] > i[1]:
            print("\nYour rent is above the median for the city which is $" + str(i[1]))
        else:
            print("\nYour rent is reasonable compared to the median of the city which is $" + str(i[1]))


## 8
thirty_per = float(format(0.3 * income, ',.2f'))
if sorted_budgets_key[7][1] >= thirty_per:
    print("\nYou are rent burdened")
    print("Your rent has to be less than $" + str(thirty_per) )
else:
    print("\nYour rent is reasonable")
```
